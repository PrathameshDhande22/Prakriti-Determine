```python
doshas=df["Dosha"].value_counts()
doshas.sort_index(inplace=True)
doshas
```

```python
target=['Vata', 'Pitta', 'Kapha', 'Vata-Pitta', 'Vata-Kapha', 'Pitta-Kapha']
target
```

```python
doshas.index=target
```

```python
doshas
```

```python
sns.barplot(doshas)
plt.xlabel("Doshas")
plt.ylabel("Counts")
plt.show()
```
